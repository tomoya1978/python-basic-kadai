```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name} は大人です")
        else:
            print(f"{self.name} は大人ではありません")

humans = [
    Human("佐藤", 25),
    Human("神谷", 18),
    Human("次郎", 20)
]

for human in humans:
    human.check_adult()
```
